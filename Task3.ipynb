```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading data 
data = pd.read_csv("SampleSuperstore.csv")
print("Data imported successfully")
```

```python
# First 5 Rows of the dataset
data.head()
```

```python
# Last 5 Rows of the dataset
data.tail()
```

```python
# Determining the shape of the dataset
print("There are total %d columns in the dataset." %data.shape[1])
print("There are total %d rows in the dataset." %data.shape[0])
```

```python
# Basic Information of the dataset
data.info()
```

```python
# Determining the Datatype of the dataset
data.dtypes
```

```python
# Determining the unique values in the dataset
data.nunique()
```

```python
#Determining the missing values in the dataset
data.isnull().sum()
```

Finding Total number of null values in a dataset

```python
print("total number of null values = ",data.isnull().sum().sum())
```

```python
#Determining the null values in the dataset
data.isna().sum()
```

```python
# Unit Price calculation
data['Unit Price'] = data.Sales / data.Quantity

# Unit Profit calculation
data['Unit Profit'] = data.Profit / data.Quantity 
```

```python
# plot of products in each category
plt.figure(figsize=(10,8))
data['Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
#checking in the Category feild.
data.groupby(["Category"]).mean()
```

```python
#checking in the Category field.
data.groupby(["Category"]).max()
```

```python
#checking in the Category field.
data.groupby(["Category"]).min()
```

```python
#Finding the totalsum and visualising the comparison of total profit with respect to the sales.
plt.figure(figsize= (10,20))
data.groupby('Category')['Profit','Sales','Unit Price','Unit Profit'].agg(['sum']).plot.bar(figsize=(15,5))
plt.title("Category vs Profit & Sales")
plt.legend(loc='best')
plt.show()
```

```python
# plot of products in each sub-category
plt.figure(figsize=(10,8))
data['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
#Checking in the sub category field
data.groupby('Sub-Category').mean()
```

```python
#Checking in the sub category field
data.groupby('Sub-Category').max()
```

```python
#Checking in the sub category field
data.groupby('Sub-Category').min()
```

```python
#for the sub category
data.groupby('Sub-Category')['Profit','Sales','Unit Price','Unit Profit'].agg(['sum']).plot.bar(figsize=(15,5))
plt.title('Total Profit, Sales, Unit Price and Unit Profit per Sub-Category', fontsize=20)
plt.ylabel('Count',fontsize=11)
plt.xlabel('Sub-Category',fontsize=11)
plt.show()
```

```python
#Count of Sub-Category region wise
plt.figure(figsize=(15,5))
sns.countplot(x="Sub-Category", hue="Region", data=data)
plt.ylabel('Count',fontsize=11)
plt.xlabel('Sub-Category',fontsize=11)
plt.title(" Sub_Category Count wrt Region",fontsize=20)
plt.show()
```

```python
# Plotting of the Correlation Matrix
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(data.corr(), annot= True)
plt.show()
```

```python
# Plotting of the Co-Variance Matrix
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(data.cov(), annot= True)
plt.show()
```

```python
# Plotting of Relation between Discount and Profit
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = data, color = 'r', label= 'Discount')
plt.ylabel('Profit',fontsize=11)
plt.xlabel('Discount',fontsize=11)
plt.title(" Discount vs Profit",fontsize=20)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10 , 6))
ax.scatter(data["Sales"] , data["Profit"])
ax.set_xlabel('Sales',fontsize=11)
ax.set_ylabel('Profit',fontsize=11)
plt.title(" Sales vs Profit",fontsize=20)
plt.show()
```

```python
# Unit Price of Sub-Category
data_subcategory = data.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit', 'Unit Price', 'Unit Profit']].mean()
data_subcategory.sort_values('Unit Price', ascending = False)[['Unit Price']].plot(kind = 'bar',figsize = (15, 5),
                                                                                 color = 'Cornflowerblue')
plt.ylabel('Average Unit Price in USD$',fontsize=11)
plt.xlabel('Sub-Category',fontsize=11)
plt.title('Average Price of Sub-Categories', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (15, 5))
# plot Sales and Profit for comparisons
sns.kdeplot(data['Sales'], color = 'Teal', label = 'Sales', shade = True, bw = 25)
sns.kdeplot(data['Profit'], color = 'Cornflowerblue', label = 'Profit', shade = True, bw = 25)
plt.xlim([0, 13000])
plt.ylim([0, 0.00007])
plt.ylabel('Density')
plt.xlabel('Monetary Value in USD$')
plt.title('Sales and Profit', fontsize = 20)
plt.legend(loc = 'best', frameon = False) 
plt.show()
```

```python
plt.figure(figsize = (15, 5))
# plot Sales and Profit for comparisons
sns.kdeplot(data['Sales'], color = 'Teal', label = 'Sales', shade = True, bw = 25)
sns.kdeplot(data['Profit'], color = 'Cornflowerblue', label = 'Profit', shade = True, bw = 25)
plt.xlim([13000, 22640])
plt.ylim([0, 0.00002])
plt.ylabel('Density')
plt.xlabel('Monetary Value in USD$')
plt.title('Sales and Profit', fontsize = 20)
plt.legend(loc = 'best', frameon = False) 
plt.show()
```

```python
#Ploting based upon the REGION and the category based.
plt.figure(figsize=(20,6))
plt.subplot(2,1,1)
sns.countplot('Region',hue='Category',data=data)
plt.title('REGION-CATEGORY')
plt.ylabel('COUNT',fontsize=11)
plt.xlabel('REGION',fontsize=11)
plt.grid(alpha=0.3)
plt.show()
```

```python
#Ploting based upon the REGION and the Sub-category based.
plt.figure(figsize=(20,6))
plt.subplot(2,1,2)
sns.countplot('Region',hue='Sub-Category',data=data)
plt.title('REGIONS-SUB CATEGORY')
plt.ylabel('COUNT',fontsize=11)
plt.xlabel('REGION',fontsize=11)
plt.grid(alpha=0.3)
plt.legend(loc='best')
plt.show()
```

```python
data.groupby('Region')['Profit'].agg(['sum']).sort_values(['sum'], ascending=False).plot.bar()
plt.show()
```

The Profit is maximum in the Western Region of the country and minimum in the Central Region of the country.

```python
data.groupby('State')['Profit'].agg(['sum']).sort_values(['sum'], ascending=False).plot.bar(figsize=(16,10))
plt.show()
```

We can see that the maximum profits comes from the State of California followed by New York. And the minimum profits comes from the States of Texas followed by Ohio and Pensylvania.
